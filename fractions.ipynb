```python
import numpy as np
import matplotlib.pyplot as plt
from jupyterthemes import jtplot

jtplot.style(theme='onedork', figsize=(16, 9))
```

```python
def fraction(x, num, den):
    return num / (den - x)
```

```python
x = np.matrix(np.linspace(-2, 2, 100)).T
num = np.array([1])
den = np.matrix(np.arange(-5, 6))

print("x: {}, num: {}, den: {}".format(x.shape, num.shape, den.shape))
```

```python
y = fraction(x, num, den)
y.shape
```

```python
plt.plot(x, y)
plt.show()
```
